```python
# 샘플 파일 만드로 지우기 
import pickle
from datetime import datetime
import os

DATA_FILE = "../data/members.dat"

# 샘플 데이터 생성
sampleData = {
    "010-2274-8999": {
        'name': '이경주',
        'relationShip': '친구',
        'address': '-',
        'regDate': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    },
    "010-1234-1234": {
        'name': '테스터',
        'relationShip': '기타',
        'address': '-',
        'regDate': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    },
    "010-0000-0001": {
        'name': '테스터2',
        'relationShip': '가족',
        'address': '서울특별시 ',
        'regDate': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    }


}

# 파일 초기화 함수
def initDataFile():
    # 파일이 존재하면 삭제
    if os.path.exists(DATA_FILE):
        os.remove(DATA_FILE)
    # 빈 파일 생성
    with open(DATA_FILE, 'wb') as f:
        pickle.dump({}, f)  # 빈 딕셔너리 저장

# 샘플 데이터 저장 함수
def makeDataFile():
    with open(DATA_FILE, 'wb') as f:
        pickle.dump(sampleData, f)

# 샘플 데이터 읽고 출력하기 
def loadDataFile():
    with open(DATA_FILE, 'rb') as f:
        data = pickle.load(f)
        
        for phone, info in data.items():
            print(f"📞 전화번호: {phone}")
            print(f"👤 이름: {info.get('name')}")
            print(f"👪 관계: {info.get('relationShip')}")
            print(f"🏠 주소: {info.get('address')}")
            print(f"🕒 등록일: {info.get('regDate')}")
            print("-" * 50)
    

# 실행 예시
# initDataFile()     # 먼저 초기화
makeDataFile()     # 샘플 데이터 저장
loadDataFile()     # 샘플 데이터 출력하기 

```
